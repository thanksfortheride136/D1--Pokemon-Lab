### Descriptive Methods
Below will contain several Pandas Descriptive Methods that are useful for sorting / categorizing / Analyzing Data.


### value_counts()
Tells you how many of each data is in a column. Use: `series.column_name.value_counts()`. In the example below the count of all the different types are shown from the "type" column, there are 112 pokemon with the "Water" type.

```python
pokemon.type.value_counts()
```

### mean()
Gets the mean of an entire column. Use: `series.column.mean()`. Access multiple columns using: `series[["column_name1", "column_name2", "column_name3"]].mean()`

-You can also compute the mean of a a Boolean Column (uses true or false instead of a number), True will map to 1 and False to 0. Therefore the % tells you what percent is true. Use: `series.column.mean() * 100`

```python
pokemon.hp.mean()
```

```python
pokemon[["hp", "attack", "defense"]]
```

```python
pokemon.legendary.mean()*100
```

### sort_values()
Allows for sorting of data. Use: `series.sort_values("column you want to sort by")`

```python
pokemon.sort_values("hp")
```

### Filtering Methods
Filtering out columns gives great flexibility in how we use data. Below are some common uses.

### Filtering a Boolean Column
To filter a Boolean column you should first create a variable that contains: `variable = series[series.column]. Simply call the variable, since they are a boolean value, it will sort to show you the "True" values by default. Note that it also shows ALL columns that correspond to the "True" in the legendary column.

```python
legendary = pokemon[pokemon.legendary]
legendary
```

### Filtering by column data
If you want to filter by content inside a column you will need to utilize a variable to hold the information and use a relational operator "==". You can also use others.
Use: `variable = series[series.column == "keyword"]`. Call the variable name to show. Use relational operators to display combinations of filtered data. See the 2nd example for use of relational operators to filter by several keywords in data.

```python
ice_pokemon = pokemon[pokemon.type == "Ice"]
ice_pokemon
```

```python
#Displays only gen1 fire pokemon
fire_gen1 = pokemon[(pokemon.generation == 1) & (pokemon.type == "Fire")] 
fire_gen1
```

```python
#Displays pokemon with a special attack greater than 70 and a speed greater than a 70 and a defense less than 90.
spAttack_speed_defense = pokemon[(pokemon.special_attack > 70) & (pokemon.speed > 70) & (pokemon.defense < 90)]
spAttack_speed_defense
```

### Grouping
Grouping allows you to group data frames together for powerful comparison abilities.

### groupby()
Allows you to group columns together. Use: `series.groupby("column").[other methods]`

1. In the first example below we are taking the mean speeds and grouping them by the type of pokemon. The mean speed of Bug pokemon is 61.6.

2. In the second example we are grouping and sorting by amount the mean amount of legendary pokemon in each generation. Generation 3 has the most legendary pokemon. Note that this can of course be saved in a variable.

3. In the third example we are grouping by the type again, but then we are finding the mean values of several columns and sorting the values by the amount of HP. Notice we need mean() for any values to be displayed.

4. In the final example we are grouping by generation, then seeing attack, special attack and defense. We are grouping by defense.

```python
pokemon.groupby("type").speed.mean()
```

```python
legendary_count_generation = pokemon.groupby("generation").legendary.mean().sort_values() * 100
legendary_count_generation 
```

```python
ptypes = pokemon.groupby("type")[["hp", "attack", "defense"]].mean().sort_values("hp")
ptypes
```

```python
gen_attackpower_defense = pokemon.groupby("generation")[["attack", "special_attack", "defense"]].mean().sort_values("defense")
gen_attackpower_defense
```

___
### More Complex Grouping (Grouping Multiple objects together)
Grouping allows you to group data together with multiple columns as well. 
1. This is showing how you would see the mean of legenday type pokemon, broken down by each type and subtype. We grouped type and subtype to get all the possible combinations, sorted by legendary mean. We also are display values that are greater than 0.
   
2. The second example is grouping by generation and type, displaying hp, attack and defense. Its being sorted by attack power.
   
3. In the third example we are doing the same thing as 2, however we are using `sort_values(by=["generation", "attack", "defense"]` to sort first by generation, then sort the individual values within attack / defense.

```python
legendary_percentages = pokemon.groupby(["type", "subtype"]).legendary.mean().sort_values() 
legendary_percentages[legendary_percentages > 0]
```

```python
attack_defense_power_over_gens = pokemon.groupby(["type", "generation"])[["attack", "defense"]].mean().sort_values("generation")
attack_defense_power_over_gens
```

```python
attack_defense_power_over_gens = (pokemon.groupby(["type", "generation"])[["attack", "defense"]].mean().sort_values(by=["generation", "attack", "defense"]))
attack_defense_power_over_gens
```

### Plotting
Pandas has good plotting capabilities, but using the seaborn library `pip3 install seaborn` has much better capabilities. 

Use: `import seaborn as sns`

Set the default theme: `sns.set_theme()`

```python
import seaborn as sns
sns.set_theme()
```


### Histograms
Good to visualize distribution of a series, puts data inot bins and plots the number of data points in each bin.

Use: `sns.histplot(data=series, xaxis_value, yaxis_value)`

```python
#plots a simple graph that counts each genrations pokemon
sns.histplot(data=pokemon, x="generation")
```

```python
#using the bins argument to increase / decrease bins
sns.histplot(data=pokemon, y="attack", bins=5)
```

### Barplot
Good when you want to compare the distribution of a numeric variable across categories, choose one numerica column and one categorical column

Use: `sns.barplot(data=series, xaxis_value, yaxis_value, errorbar="sd"[optional])`

```python
sns.barplot(data=pokemon, x="special_attack", y="type", errorbar=None, palette="muted")
```

```python
sns.barplot(data=pokemon, x="attack", y="type", errorbar=None)
```

### Scatterplot
Good to show a relationship between 2 numeric variables. Looking at a point in the example below gives you the y value or speed of a pokemon, tracing it down to the x value or attack gives the attack. Together 1 point will display those 2 values.

Use: `sns.scatterplot(data=series, xvalue, yvalue)`

In the 2nd we are distuingishing categories within the scatter plot using hue. You would want something with a small amount of options. In the pokemon data set I used generation for this, it has 6 options which may be a lot.

Use: `sns.scatterplot(data=series, x="column_name", y="column_name", hue="column_name", alpha=int, s=int)`



```python
sns.scatterplot(data=pokemon, x="attack", y="speed")
```

```python
sns.scatterplot(data=pokemon, x="defense", y="special_attack", hue="generation", alpha=0.5, s=60)
```

### relplot()
Lets you plot across multiple categories and make several plots at once. 

You use an x and y parameter as usual, but there also is a `col` paraemeter which allows you to add different columns of charts. In the first example we are plotting the attack and speed variables, we are using `col="legendary"` to see how legendary vs non legendary pokemon's attack and speed values relate.

Use: `sns.relplot(data=series, x="xvalue", y="yvalue", col="another_column")`

In the 2nd example we are accessing values inside a column, instead of a full column.




```python
sns.relplot(data=pokemon, x="speed", y="attack", col="legendary")
```

```python
fav_types = pokemon[pokemon.type.isin(["Fighting", "Ghost", "Grass"])]
sns.relplot(data=fav_types, x="hp", y="speed", hue="legendary", col="type", s=100)
```
